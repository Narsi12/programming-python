```python
import requests

# get call
def get_data(id_num:int,base_url):
    url = f'{base_url}/{id_num}'
    response = requests.get(url)
    return response.json()
base_url  = 'https://jsonplaceholder.typicode.com/posts'
id_num = 1
print(get_data(id_num,base_url))

# post call

def post_data(title:str,body:str,user_id:int):
    url = 'https://jsonplaceholder.typicode.com/posts'
    payload = {
        "title":title,
        "body":body,
        "user_id":user_id
    }
    response = requests.post(url,json = payload)
    return response.json()


title = "mouritech"
body = "Mid level MNC"
user_id = 24
print(post_data(title,body,user_id))
```

```python
import requests

# get call
def get_data(id_num:int,base_url):
    url = f'{base_url}/{id_num}'
    response = requests.get(url)
    data = response.json()
    if data['userId'] == 1:
        return data
    else:
        return "No record found"
base_url  = 'https://jsonplaceholder.typicode.com/posts'
id_num = 1
print(get_data(id_num,base_url))
```

```python
import asyncio
import aiohttp

count = 0

# async GET
async def get_data(session, id_num: int, base_url: str):
    global count
    count+=1
    url = f"{base_url}/{id_num}"
    async with session.get(url) as response: # session here is used for http calls like get, post
        return await response.json() # await = wait without blocking

# run 20 calls in parallel
async def fetch_all():
    base_url = "https://jsonplaceholder.typicode.com/posts"
    
    async with aiohttp.ClientSession() as session: # insted of requests.get() we used  aiohttp.ClientSession() , ClientSession() = A reusable HTTP client (connection manager)
        tasks = [get_data(session, i, base_url) for i in range(1, 21)]
        results = await asyncio.gather(*tasks)
        
        return results

results = await fetch_all()
print(results)
print(f"Total API Calls {count}")
```

```python
import numpy as np

def cosine_similarity(vec_a, vec_b):
    # Measures the angle between two vectors, not their distance
    dot_product = np.dot(vec_a, vec_b)
    magnitude = np.linalg.norm(vec_a) * np.linalg.norm(vec_b)
    return dot_product / magnitude
    # Returns 1.0 = identical meaning, 0.0 = unrelated, -1.0 = opposite

# In practice with real embeddings:
from openai import OpenAI
client = OpenAI()

def get_embedding(text):
    response = client.embeddings.create(
        input=text,
        model="text-embedding-3-small"
    )
    return response.data[0].embedding  # a list of 1536 floats

vec1 = get_embedding("The dog ate my homework")
vec2 = get_embedding("A canine consumed my assignment")
vec3 = get_embedding("The stock market crashed")

print(cosine_similarity(vec1, vec2))  # ~0.94 — very similar
print(cosine_similarity(vec1, vec3))  # ~0.12 — very different
```
