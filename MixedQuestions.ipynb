```python
#1. flatten list
def flatten(data):
    flat=[]
    for item in data:
        if isinstance(item,list):
            flat.extend(flatten(item))
        else:
            flat.append(item)
    return flat
            
data = [1, [2, 3], [4, [5, 6]], 7]
print(flatten(data))
```

```python
# 5 . group of anagrams
words = ["eat","tea","tan","ate","nat","bat"]
dic = {}
for i in words:
    key = ''.join(sorted(i))
    if key not in dic:
        dic[key] = [i]
    else:
        dic[key]+=[i]
print(list(dic.values()))
```

```python
# 8. Check Balanced Parentheses
def valid_params(s):
    stack = []
    mapping = {')':'(',']':'[','}':'{'}
    for item in s:
        if item in mapping:
            top_ele = stack.pop() if stack else '#'
            if mapping[item] != top_ele:
                return False
        else:
            stack.append(item)
    return not stack

s = "{[()]}"
print( valid_params(s) )
            
        
```

```python
# 9. flatten dict
def flatten_dict(d,parent=''):
    res = {}
    for key,val in d.items():
        new_key = parent+key
        if isinstance(val,dict):
            res.update(flatten_dict(val,new_key))
        else:
            res[new_key]=val
    return res
         
d = {'a': {'b': {'c': 1}}, 'd': 2}
print(flatten_dict(d))
```

```python
# 12 Minimum window substring (basic idea)
def solution(string, pat):
    left = 0
    start = 0
    count = len(pat)
    min_len = float('inf')
    need = {}
    for i in pat:
        need[i] = need.get(i,0)+1
        
    for right in range(len(string)):
        if string[right] in need:
            if need[string[right]]>0:
                count-=1
            need[string[right]]-=1
            
        while count == 0:
            if right-left+1<min_len:
                min_len = right-left+1
                start=left
                
            if string[left] in need:
                need[string[left]]+=1
                if need[string[left]]>0:
                    count+=1
            left+=1
    return string[start:start+min_len]

string = "ADOBECODEBANC"
pat = "ABC"
print(solution(string, pat))
```

```python
# 14. create a iterator
class Iteration:
    def __init__(self,start,end):
        self.start = start
        self.end = end
    def __iter__(self):
        return self
    def __next__(self):
        if self.start< self.end:
            self.start+=1
            return self.start-1
        else:
            raise StopIteration
my = Iteration(20,30)
print(next(my))
print(next(my))
print(next(my))
print(next(my))
```

```python
# 15. write a simple generator
def my_gen(start,end):
    while start<end:
        yield start
        start+=1
x = my_gen(10,20)
print(next(x))
print(next(x))
print(next(x))
```

```python
# 16. find the second largest
def second_large(nums):
    first=second=float('-inf')
    for num in nums:
        if num>first:
            second = first
            first = num
        elif second > num and num!=first:
            second = num
    return second

print(second_large([10, 20, 4, 45, 99,]))
```

```python
#17. factorial of a number
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact = fact*i
    return fact
print(factorial(5))
```

```python
# 18 . fibbonaci series
def fibbo(n):
    fib_series = [0,1]
    for i in range(2,n+1):
        fib_series.append(fib_series[i-1]+fib_series[i-2])
    return fib_series

print(fibbo(10))
```

```python
# 21. merge intervals
def merge_intervals(data):
    data.sort(key=lambda x:x[0])
    lst_block = [data[0]]
    for start,end in data[1:]:
        lst_start,lst_end = lst_block[-1]
        if lst_end > start:
            lst_block[-1][1] = max(end,lst_end)
        else:
            lst_block.append([start,end])
    return lst_block
    
intervels = [[1,3],[2,4],[5,7],[6,8]]
print(merge_intervals(intervels))
```
