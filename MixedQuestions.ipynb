```python
#1. flatten list
def flatten(data):
    flat=[]
    for item in data:
        if isinstance(item,list):
            flat.extend(flatten(item))
        else:
            flat.append(item)
    return flat
            
data = [1, [2, 3], [4, [5, 6]], 7]
print(flatten(data))
```

```python
# 3. merge two dicts
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}
c = {}
for k in d1:
    c[k] = d1[k]
for k in d2:
    if k in c:
        c[k]+=d2[k]
    else:
        c[k] = d2[k]
c
 
```

```python
# 4.find duplicates
arr = [1,2,3,2,4,1,5]
dup = []
seen = set()
for i in arr:
    if i not in seen:
        seen.add(i)
    else:
        dup.append(i)
seen
        
 
```

```python
# 5 . group of anagrams
words = ["eat","tea","tan","ate","nat","bat"]
dic = {}
for i in words:
    key = ''.join(sorted(i))
    if key not in dic:
        dic[key] = [i]
    else:
        dic[key]+=[i]
print(list(dic.values()))
```

```python
# 8. Check Balanced Parentheses
def valid_params(s):
    stack = []
    mapping = {')':'(',']':'[','}':'{'}
    for item in s:
        if item in mapping:
            top_ele = stack.pop() if stack else '#'
            if mapping[item] != top_ele:
                return False
        else:
            stack.append(item)
    return not stack

s = "{[()]}"
print( valid_params(s) )
            
        
```
