```python
# simple left handle triangle
n = 5
for i in range(1,n+1):
    for j in range(i):
        print("*",end=' ')
    print()
```

```python
# print right hand triangle
n = 5
for i in range(n+1):
    for j in range(n-i): # or (i,n)
        print("*",end=' ')
    print()
```

```python
# print pyramid joined triangle
n = 5
for i in range(1,n+1):
    for j in range(n-i):
        print(" ",end=' ')
    for k in range(1,2*i):
        print("*",end=' ')
    print()
```

```python
# right angle tri-angle
n = 6
for row in range(1,n+1):
    for i in range(n-row):
        print(" ",end=' ')
    for j in range(row):
        print("*",end=' ')
    print()
```

```python
# leftb tri-angle
n = 6
for row in range(1,n+1):
    for i in range(row):
        print(" ",end=' ')
    for j in range(n+1-row): # +1 added for size big you can remove and increase the n = 8
        print("*",end=' ')
    print()
```

```python
# new pyramid
n=5
for row in range(n+1):
    for j in range(n-row):
        print(" ",end=' ')
    for k in range(row):
        print("*",end=' ')
    for p in range(row-1):
        print("*",end=' ')
    print()

 
```

```python
 # new reverse pyramid
n=5
for row in range(n+1):
    for j in range(row+1):
        print(" ",end=' ')
    for k in range(row,n):
        print("*",end=' ')
    for l in range(row+1,n):
        print("*",end=' ')
    print()
```

```python
# diffrenst approach insted of 3 for loops using two only
n = 5
for i in range(n+1):
    for j in range(n-i):
        print(" ",end =" ")
    for j in range(1,i*2):
        print("*",end= " ")
    print()
    
```
