```python
# 12. sort the dict by keys
d={'b':2,'a':1}
c = list(d.items())
for i in range(len(c)):
    for j in range(i+1,len(c)):
        if c[i][0] > c[j][0]:
            c[i],c[j]=c[j],c[i]
print(dict(c))
```

```python
# 13. Sort dict by values
d={'a':3,'b':1}
data = dict(sorted(d.items(), key=lambda x:x[1]))
data
```

```python
# 14. Invert dictionary
d={'a':1,'b':2}
inv = {}
for k,v in d.items():
    inv[v]=k
inv
```

```python
# 15.Handle duplicate values while invert
d={'a':1,'b':1}
dup = {}
for k,v in d.items():
    if v not in dup:
        dup[v] = [k]
    else:
        dup[v].append(k)
dup
    
```

```python
# 16. Find max value key
d={'a':10,'b':20}
maxy = float('-inf')
high_key = None
for k, v in d.items():
    if v>maxy:
        maxy = v
        high_key = k
high_key
        
```

```python
# 20. Flatten dict
def flatten(d):
    flat = {}
    for k,v in d.items():
        if type(v) == dict:
            for k2,v2 in v.items():
                flat[k+'.'+k2]=v2
        else:
            flat[k]=v
    return flat

d = {'a': 1, 'b': {'c': 2, 'd': 3}}
print(flatten(d))
```
