```python
#2. Valid parthesesis
def valid_parthesis(s):
    stack =[]
    pairs = {')':'(',']':'[','}':'{'}
    for ch in s:
        if ch in pairs:
            top_ele = stack.pop() if stack else '#'
            if pairs[ch] != top_ele:
                return False
        else:
            stack.append(ch)
    return not stack
s = "[{()}]" 
print(valid_parthesis(s))
```

```python
# 3. Merge two sorted list

def merge_list(a,b):
    i=0
    j=0
    res = []
    while i < len(a) and j < len(b):
        if a[i]<b[j]:
            res.append(a[i])
            i+=1
        else:
            res.append(b[j])
            j+=1
    res.extend(a[i:])
    res.extend(b[j:])
    return res
            

a = [1,3,5]
b = [2,4,6]
print(merge_list(a,b))
```

```python
# 4 . Remove Duplicates from Array
def remove_dups(arr):
    i = 0
    for j in range(1,len(arr)):
        if arr[i]!=arr[j]:
            i+=1
            arr[i]=arr[j]
    return arr[:i+1]

arr = [1,1,2,2,3,4,4]
print(remove_dups(arr))
```

```python
# 5. Find Missing Number
def find_miss_num(arr):
    n = len(arr)
    exp_sum = n*(n+1)//2
    ac_sum = sum(arr)
    return ac_sum - exp_sum

arr = [1,3,4]
print(find_miss_num(arr))
```

```python
# 6. Move Zeros to End
def move_zeros_end(arr):
    i = 0
    for j in range(len(arr)):
        if arr[j]!=0:
            arr[i],arr[j]=arr[j],arr[i]
            i+=1
    return arr

arr = [1,0,22,4,0,5,40,0,6,0]
print(move_zeros_end(arr))
```

```python
# 7. heck Palindrome String
def string_palindrome(s):
    s = s.lower()
    l = 0
    r = len(s)-1
    while l < r:
        if s[l]!=s[r]:
            return False
        l+=1
        r-=1
    return True

s = "Racecar"
print(string_palindrome(s))
```

```python
# 8. Fibonacci Series
def fibbo_series(n):
    a=0
    b=1
    for _ in range(n):
        print(a,end=" ")
        a,b=b,a+b
n = 7
print(fibbo_series(n))
```
