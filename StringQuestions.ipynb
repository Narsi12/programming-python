```python
# 6. check anagrams

def check_anagrams(a,b):
    count = {}
    if len(a) != len(b):
        return False
    for ch in a:
        if ch not in count:
            count[ch]=1
        else:
            count[ch]+=1
    for ch in b:
        if ch not in count or count[ch]==0:
            return False
        count[ch]-=1
    return True
    

a = "listen"
b = "silent"
print(check_anagrams(a,b))
 
```

```python
# 14. Check valid parentheses
def valid_params(data):
    stack = []
    mapping = {')':'(',']':'[','}':'{'}
    for char in data:
        if char in mapping:
            top_ele = stack.pop() if stack else "#"
            if mapping[char] != top_ele:
                return False
        else:
            stack.append(char)
    return not stack

data = "({})[]{}"
print(valid_params(data))
```

```python
# 17. Minimum window substring (basic idea)

def solution(string, pat):
    start = 0
    left = 0
    count = len(pat)
    need = {}
    min_len = float('inf')
    for i in pat:
        need[i] = need.get(i,0)+1
    for right in range(len(string)):
        if string[right] in need:
            if need[string[right]]>0:
                count-=1
            need[string[right]]-=1
            
        while count == 0:
            if right - left+1 < min_len:
                min_len = right - left+1
                start = left
            if string[left] in need:
                need[string[left]]+=1
                if need[string[left]]>0:
                    count+=1
            left+=1
    return string[start:start+min_len]
    
string = "ADOBECODEBANC"
pat = "ABC"
print(solution(string, pat))


    
```
