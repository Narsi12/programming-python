```python
# 2 . check palindrome
word = "Madam"
word = word.lower()
l = 0
r = len(word)-1
while l<r:
    if word[l] != word[r]:
        print("NO")
        break
    l+=1
    r-=1
else:
    print("YES")
```

```python
# 3. count vowels
word = "hello world"
vowels = "aeiou"
count = 0
for i in word:
    if i in vowels:
        count+=1
print(count)
```

```python
# 6. check anagrams

def check_anagrams(a,b):
    count = {}
    if len(a) != len(b):
        return False
    for ch in a:
        if ch not in count:
            count[ch]=1
        else:
            count[ch]+=1
    for ch in b:
        if ch not in count and count[ch]==0:
            return False
        count[ch]-=1
    return True
    

a = "listen"
b = "silent"
print(check_anagrams(a,b))
 
```
