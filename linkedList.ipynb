A Linked List is a linear data structure where elements (called nodes) are stored in non-contiguous memory locations.
Each node contains:

Data – the value stored

Pointer (next) – address/reference of the next node in the list

Nodes are connected like a chain, forming a sequence.

# singly linked list
[Data | Next] -> [Data | Next] -> [Data | Next] -> None



```python
# create a simple linkedlist with three nodes that cantaines address of another nodes
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
node1 = Node(10)
node2 = Node(15)
node3 = Node(25)
node4 = Node(76)

node1.next = node2
node2.next = node3
node3.next = node4

head = node1
current = head
while current:
    print(current.data,end=" -> ")
    current = current.next
print("None")
```

```python
# create a nodes dynamically
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
def create_a_nodes(values):
    if not values:
        return None
    head = Node(values[0])
    curr = head
    for val in values[1:]:
        curr.next = Node(val)
        curr = curr.next
    return head
head = create_a_nodes([10,20,30,40,50])
current = head
while current:
    print(current.data,end=" -> ")
    current = current.next
print("None")
```

```python
# find the number of nodes in linkedlist
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def create_a_node(values):
    if not values:
        return None
    head = Node(values[0])
    current = head
    for val in values[1:]:
        current.next=Node(val)
        current = current.next
    return head


def count_nodes(head):
    count = 0
    current = head
    while current is not None:
        count+=1
        current = current.next
    return count


head = create_a_node([10,20,30,40])
print("Number of nodes", count_nodes(head))
current = head
while current:
    print(current.data,end=" -> ")
    current = current.next
print("None")
```
